```python
import torch.nn as nn

class FrodClassifier(nn.Module):
    def __init__(self):
        super(FrodClassifier, self).__init__()
        self.fc1 = nn.Linear(1000, 500)
        self.relu = nn.ReLU()
        self.sig = nn.Sigmoid()
        self.fc2 = nn.Linear(500, 100)
        # self.relu = nn.ReLU()
        # self.sig = nn.Sigmoid()
        self.fc3 = nn.Linear(412, 100)
        # self.relu = nn.ReLU()
        self.fc4 = nn.Linear(100, 2)

    def forward(self, x):
        cuda = torch.device('cuda')

        x = x.tolist()
        res1 = []
        txt = []
        for el in x:
          res1.append(el[:1000])
          txt.append(el[1000:])

        res = torch.Tensor(res1).float().to(cuda)
        res = self.fc1(res)
        res = self.relu(res)
        res = self.fc2(res)
        res = self.relu(res)

        res = res.tolist()
        all = []
        for i in range(len(res)):
          all.append(res[i] + txt[i])
        all = torch.Tensor(all).float().to(cuda)

        all = self.fc3(all)
        all = self.relu(all)
        all = self.fc4(all)
        return all
```

```python
if torch.cuda.is_available():
  device = 'cuda'
else:
  device = 'cpu'
print(device)
```

```python
model = FrodClassifier().to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
loss_fn = torch.nn.CrossEntropyLoss()
```

```python
class Dataset(torch.utils.data.Dataset):
    def __init__(self, embeddings, labels):
        super(Dataset, self).__init__()
        self.embeddings = embeddings
        self.labels = labels
    def __len__(self):
        return self.embeddings.shape[0]

    def __getitem__(self, idx):
        return ( np.array(self.embeddings.iloc[idx]), np.array(self.labels.iloc[idx]))
```

```python
train_dataset, test_dataset = Dataset(X_train, y_train), Dataset(X_test, y_test)
```

```python
len(train_dataset)
```

```python
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=64, shuffle=False)
```

```python
next(iter(train_loader))
```

```python
import matplotlib.pyplot as plt
from IPython.display import clear_output
```

```python
n_epoch = 16
train_losses, val_losses = [], []
train_acc, val_acc = [], []
for epoch in range(n_epoch):
    clear_output(wait=True)
    print("Epoch:", epoch)

    # отрисовка графиков обучения
    plt.figure(figsize=(15, 3))
    plt.subplot(1, 4, 1)
    plt.ylim([0, 1])
    plt.title('Train loss')
    plt.plot(train_losses)
    plt.grid()

    plt.subplot(1, 4, 2)
    plt.title('Val loss')
    plt.ylim([0, 1])
    plt.plot(val_losses)
    plt.grid()

    plt.subplot(1, 4, 3)
    plt.title('Train accuracy')
    plt.ylim([0, 1])
    plt.plot(train_acc)
    plt.grid()

    plt.subplot(1, 4, 4)
    plt.title('Val accuracy')
    plt.ylim([0, 1])
    plt.plot(val_acc)
    plt.grid()
    plt.show()

    model.train()
    losses = []
    acc = []
    for i, data in tqdm(enumerate(train_loader)):
        inputs, labels = data

        optimizer.zero_grad()

        outputs = model(inputs.to(device, dtype=torch.float32))
        assert not torch.isnan(outputs).sum()
        loss = loss_fn(outputs, labels.to(device))
        loss.backward()
        optimizer.step()
        losses.append(loss.item())
        acc.append(((labels.to(device) == outputs.argmax(dim=1)).sum() / len(labels)).cpu())

    train_losses.append(np.mean(losses))
    train_acc.append(np.mean(acc))

    model.eval()
    losses = []
    acc = []
    with torch.no_grad():
        for i, vdata in enumerate(test_loader):
            inputs, labels = data

            outputs = model(inputs.to(device, dtype=torch.float32))

            loss = loss_fn(outputs, labels.to(device))

            losses.append(loss.item())
            acc.append(((labels.to(device) == outputs.argmax(dim=1)).sum() / len(labels)).cpu())

    val_losses.append(np.mean(losses))
    val_acc.append(np.mean(acc))
```
